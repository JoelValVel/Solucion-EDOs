```python
from ODEsolver import ODEsolver
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import RMSprop, Adam
import matplotlib.pyplot as plt
import numpy as np
```

```python
3*tf.math.sin(1*np.math.pi)
```

```python
minval=-1
maxval = 1
ic11 = 3*tf.math.sin(-1*np.math.pi)
ic0 = 3*tf.math.sin(0*np.math.pi)
ic1 = 3*tf.math.sin(1*np.math.pi)
model = ODEsolver(funcion="3*tf.math.sin(x*np.math.pi)", ic11=2.6226832e-07, ic0=0.0,ic1=-2.6226832e-07)

model.add(Dense(10, activation='tanh', input_shape=(1,)))
model.add(Dense(1, activation='tanh'))
model.add(Dense(1, activation='linear'))

model.summary()

model.compile(optimizer=Adam(), metrics=['loss'])

x = tf.linspace(minval,maxval,1000)
history = model.fit(x, epochs=2000, verbose=1)

a = model.predict(x)
plt.plot(x,a, label="Solución numérica")
plt.plot(x, 3*tf.math.sin(x*np.math.pi),label="Solución analítica")
plt.legend()
plt.show()
```

```python
minval=-1
maxval = 1
# ic11: valor de la función en x=-1
# ic0: valor de la función en x=0
# ic1: valor de la función en x=1
model = ODEsolver(funcion="1+2*x+4*x**3",ic11=-5, ic0=1,ic1=7)

model.add(Dense(10, activation='tanh', input_shape=(1,)))
model.add(Dense(1, activation='tanh'))
model.add(Dense(1, activation='linear'))

model.summary()

model.compile(optimizer=Adam(), metrics=['loss'])

x = tf.linspace(minval,maxval,1000)
history = model.fit(x, epochs=2000, verbose=1)

a = model.predict(x)
plt.plot(x,a, label="Solución numérica")
plt.plot(x, 1+2*x+4*x**3,label="Solución analítica")
plt.legend()
plt.show()
```
